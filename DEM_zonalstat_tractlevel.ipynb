```python
import ee
import geemap
import geemap.colormaps as cm
```

## County level DEM

```python
Map = geemap.Map(center=[40, -100], zoom=4)
Map
dem = ee.Image('USGS/SRTMGL1_003')
vis = {'min': 0, 'max': 4000, 'palette': cm.palettes.dem}
Map.addLayer(dem, vis, 'DEM')

#county = ee.Image("TIGER/2016/Counties") #.filter(ee.Filter.eq('STATEFP', '23'));
county = ee.FeatureCollection("TIGER/2018/Counties") #.filter(ee.Filter.eq('STATEFP', '23'));
Map.addLayer(county, {}, 'US States')
Map
out_dem_stats = './dem_stats.csv'
geemap.zonal_statistics(dem, county, out_dem_stats, statistics_type='MEAN', scale=2000)
```

# census tract level DEM

```python
import pandas as pd
```

```python
DF = pd.read_csv('./StateFIPS.csv')
DF['FIPS'] =DF['FIPS'].apply(lambda x: '{0:0>2}'.format(x))
DF
```

```python
dem = ee.Image('USGS/SRTMGL1_003')
#county = ee.Image("TIGER/2016/Counties") #.filter(ee.Filter.eq('STATEFP', '23'));
for fips in DF['FIPS']:
    county = ee.FeatureCollection("TIGER/2020/TRACT").filter(ee.Filter.eq('STATEFP', fips));
    out_dem_stats = './dem_stats'+fips+'.csv'
    geemap.zonal_statistics(dem, county, out_dem_stats, statistics_type='MEAN', scale=2000)
```
